### Extract the URLs from the item page
An example item page now is http://www.hm.com/us/products/ladies?page=1, which contains only 60 items.

The URLs can be obtained from __'a' tag with class "product-url js-product-tracking"__. There are only 60 tags withs __"product-url js-product-tracking"__.

```python
import urllib
```

```python
print url_women
```

```python
handle = urllib.urlopen(url_women)
```

```python
html_gunk =  handle.read()
```

```python
from HTMLParser import HTMLParser
```

```python
class HnM_ItemList_Parser(HTMLParser):
    in_item_url_tag = False
    item_url_list = []
    def handle_starttag(self, tag, attrs):
        if tag == "a":
            for attr in attrs:
                if attr[0] == "class" and attr[1] == "product-url js-product-tracking":
                    self.in_item_url_tag = True
                    break
            if self.in_item_url_tag:
                for attr in attrs:
                    if attr[0] == "href":
                        self.item_url_list.append(str(attr[1]))
    def get_item_url_list(self):
        return self.item_url_list[:]
```

```python
parser = HnM_ItemList_Parser()
```
