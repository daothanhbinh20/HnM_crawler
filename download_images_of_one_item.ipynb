```python
list_image_name = []
```

```python
for key in items_info[0]['image']:
    image_name = items_info[0]['who'][0].upper() + key[0].upper() + '_' + items_info[0]['serial']
    list_image_name.append(image_name)
```

```python
print list_image_name
```

* WI: Women Item image
* WM: Women Model image
* MI: Men Item image
* MM: Men Model image

```python
image_folder = 'item_images'
```

```python
list_image_path = []
```

```python
for image_name in list_image_name:
    image_path = image_folder + '/' + image_name + '.jpg'
    list_image_path.append(image_path)
```

```python
list_image_path
```

```python
import urllib

def save_image(path, url_image):
    image_contents = urllib.urlopen(url_image)
    f = open(path,'wb')
    f.write(image_contents.read())
    image_contents.close()
    f.close()
```

```python
save_image(list_image_path[0], items_info[0]['image']['item'])
save_image(list_image_path[1], items_info[0]['image']['model'])
```
